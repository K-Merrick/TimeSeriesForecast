### Examine Seasonality, Trends, and Residuals

```python
#Create function to plot trend, seasonality, and residuals.
def tsr_decompose(df):
    """Decomposes time series data and creates plots for trend, seasonality, and residuals."""

    #Decompose data.
    decomposed = seasonal_decompose(df)

    #Create figure.
    fig = plt.figure(figsize=(13, 8))
    fig.suptitle("Decomposed Data", fontsize=15, x=.515)
    
    #Create subplots
    plt.subplot(311)
    decomposed.trend.plot(ax=plt.gca())
    plt.title('Trend', fontsize=12) ; plt.xlabel('')
    plt.subplot(312)
    decomposed.seasonal.plot(ax=plt.gca())
    plt.title('Seasonal', fontsize=12) ; plt.xlabel('')
    plt.subplot(313)
    decomposed.resid.plot(ax=plt.gca())
    plt.title('Residuals', fontsize=12) ; plt.xlabel('')
    plt.tight_layout()
```

### Check for Stationarity

A stationary time series in one whose properties don't depend on the times at which the series is observed. For example, if a variable's value is correlated with seasons or trends, then the series is non-stationary, as both seasons and trends affect the value of the series at different times. In order to build a model that forecasts values based on a time series, the series must be stationary. In this subsection we will use two methods to test for stationarity. First, we will plot the series along with its rolling mean and standard deviation - if these measures do not change, then the series is stationary. Next, we will run an Augmented Dickey-Fuller test, which essentially tests the hypothesis that the time series is non-stationary by examining how strongly a time series is affected by a trend. If we find that our time series is non-stationary, then we will transform the data in such a way as to make it stationary.

#### Examine Plot for Stationarity

```python
#Create function to plot data with rolling mean and window. 
def stationarity_check_plot(df, mean_window, sd_window):
    """ This function plots the values of a time series, along with a rolling mean and standard deviation.
    
        Arguments include:
        
        df - Series to be plotted.
        mean_window - The size of the rolling mean window.
        sd_window - The size of the rolling standard deviation window.
        """
        
    #Create plot figure.
    fig = plt.figure(figsize=(15, 7))
    fig.suptitle('Time Series Data with Rolling Mean and Standard Deviation', fontsize=15, y=.92)
        
    #Plot data.
    sns.lineplot(x='datetime', y='num_orders', data=df)
    sns.lineplot(x='datetime', y='num_orders', data=df.rolling(mean_window).mean())
    sns.lineplot(x='datetime', y='num_orders', data=df.rolling(sd_window).std())
    plt.ylabel('Number of Orders', fontsize=12, labelpad=10)
    plt.xlabel('Date/Time of Observation', fontsize=12, labelpad=10)
    plt.legend(labels=['Original Data', f'Rolling Mean ({mean_window})', f'Rolling Standard Deviation ({sd_window})'])
    plt.show()
```

```python
#Plot data for time series.
#Use window of 24h to represent changes by day.
stationarity_check_plot(df, 24, 24)
```

2.1.1 Notes: Based on our graph, it appears that the time series is not stationary - that is, there appears to be a slight upwards trend in the number of taxi orders, as well as the rolling mean and standard deviation. In order to ensure that our interpretation of the graph is correct, we will run an Augmented Dickey-Fuller test.

#### Run Augmented Dickey-Fuller Test

An Augmented Dickey-Fuller (ADF) test is one type of 'Unit Root' test. Such tests run a statistical analysis on a series to find out if there is a unit root - that is, if the coefficient for the first lag of a time series is equal to 1. 

The ADF test essentially tests the null hypothesis that the time series is non-stationary. The alternate hypothesis in this case is that the series is stationary. If the resultant test statistic of an ADF test is less than the critical values and the p-value is less than the thresholds, then we reject the null hypothesis. If these values are greater than their relevant counterparts, then we fail to reject the null hypothesis.

```python
#Create function to run Augmented Dickey-Fuller test and get values.
def adfuller_test(df):
    #Run test.
    adf_results = adfuller(df)
    #Print pertinent test values.
    print('ADF Statistic: {}'.format(adf_results[0]))
    print('p-value: {}'.format(adf_results[1]))
    print('Critical Values:')
    for key, value in adf_results[4].items():
        print('\t{}: {}'.format(key, value))
        
    if adf_results[0] < list(adf_results[4].values())[0] and adf_results[1] < .01:
        print('\nWe reject the (null) hypothesis that the time series is non-stationary.')
    else:
        print('\nWe fail to reject the null hypothesis.')
```

```python
#Run ADF test on time series.
print('Results of ADF test on time series: \n')
adfuller_test(df['num_orders'])
```

2.2.2 Notes: Per the results of the ADF test on our time series, we fail to reject the null hypothesis that the series is non-stationary. This supports our intuition from the previous subsection, based on the slightly increasing rolling mean and standard deviation, that the series is non-stationary. In order to carry out forecasting, we will attempt to make the series stationary by differencing it; this method should remove seasonality and trend from the series. 

In the following subsection, we will perform this differencing technique on our time series and test for stationarity.

### Make Time Series Stationary

#### Perform Differencing Technique

As mentioned, the differencing technique is one way to remove trend and seasonality from a time series. Regarding the latter, we saw that one seasonal cycle is typically one day. In order to remove this seasonality from our series, we will perform differencing using a lag value of 24 (since there are 24 hours in a day). We will then plot the new series, decompose it, and run an ADF test to see if our series is stationary.

```python
#Create new series by differencing with 1 shift. 
df_diff = df - df.shift(24)

#Drop NaN in first row.
df_diff.dropna(inplace=True)

#Display first 5 rows of time series.
display(df_diff.head())
```

#### Examine New Series

```python
#Plot data for time series.
#Use window of 24h to represent changes by day.
stationarity_check_plot(df_diff, 24, 24)
```

2.3.2 Notes: Here we see that the trends present in subsection 2.2.1 are no longer present in our differences series. That is, while there is some oscillation in the rolling mean and standard deviation, they roughly oscillate around the same value. While there may appear to be a very slight increase in the general rolling standard deviation towards September, this may be insignificant with regard to our later modeling. To see if we've removed much of the trend and seasonality from our data, let us examine the decomposed plots.

```python
#Decompose the data and plot trend, seasonality, and residuals.
tsr_decompose(df_diff)
```

2.3.2 Notes Continued: From these plots we can see that we've greatly reduced the effects of trend and seasonality on our series. With regard to the former, the trend values appear to oscillate around 0, and the overall trend does not appear to increase or decrease. Regarding seasonality, the range of values is between -.5 and 1, which seems insignificant. In order to statistically test whether our series is likely stationary, we will run an ADF test in the next subsection.

#### Run Augmented Dickey-Fuller Test on Differenced Data

```python
print('Results of ADF test on time series: \n')
adfuller_test(df_diff['num_orders'])
```

2.3.3 Notes: Based on the results of the ADF test, we have good support for the proposition that our time series is now stationary. Not only is the ADF statistic below the critical values, but also the p-value is far below the 1% threshold; regarding the latter, it means that we can reject the null hypothesis with a confidence of greater than 99%. 

Assuming that our time series is now stationary, we will move on to check whether there is significant autocorrelation.

### Check for Autocorrelation

Autocorrelation is a type of serial dependence; it refers to instances in which a time series is linearly related to a lagged version of itself. When there is a high correlation between a time series and some lagged version of itself, the series is said to be autocorrelated. Whereas autocorrelation can be used to determine whether there is seasonality in a time series, it is problematic for forecasting; this is because many forecasting methods rely on the assumption that there is no autocorrelation in the residuals. In this subsection, we will look at the Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) plots. The former pertains to the correlation between a series and a lagged versions of itself; the latter pertains to the same, though where each correlation is controlled for any correlation between the series and lag lengths shorter than the lag value being examined. If there is significant autocorrelation, then we would have to correct for this by transforming the data even more. 

#### Examine ACF and PACF for Original Series

```python
#Create function to neatly display plots.
def p_acf(df, lag):
    #Create figure.
    fig = plt.figure(figsize=(15, 10))
    fig.suptitle("Original, ACF, and PACF Plots", fontsize=15, y=.91)
    #Create subplot grids.
    orig = plt.subplot2grid((2, 2), (0, 0), colspan = 2)
    acf_plt = plt.subplot2grid((2, 2), (1, 0))
    pacf_plt = plt.subplot2grid((2, 2), (1, 1))
    #Create original data plot.
    df.plot(ax=orig)
    df.rolling(24).mean().plot(ax=orig)
    df.rolling(24).std().plot(ax=orig)
    orig.legend(labels=['Original Data', f'Rolling Mean ({24})', f'Rolling Standard Deviation ({24})'])
    orig.set_xlabel('')
    #Use ACF and PACF plots into subplots.
    plot_acf(df['num_orders'], lags=lag, ax=acf_plt)
    plot_pacf(df['num_orders'], lags=lag, ax=pacf_plt)
    plt.show()
```

```python
#Plot data for original series.
p_acf(df, 48)
```

2.4.1 Notes: As we can see,the both the ACF and PACF graphs indicate that there is a substantial number of lags in which there is statistically significant autocorrelation. On method of correcting for this is transforming the data using differencing, which we have done. In the next subsection, we will examine the plots for our differenced data; if there is still a significant number of lags with high autocorrelation, then we may have to difference the data once more, as it would indicate that there is still trend or seasonality within our data.

#### Examine ACF and PACF Plots for Differenced Series

```python
#Plot data for differenced series.
p_acf(df_diff, 48)
```

2.4.2 Notes: Here we see that our differenced series has far lower autocorrelation and partial autocorrelation than our original data. Where there are a number of peaks outside of the threshold in both plots, the peaks are far lower than before. Given this weaker correlation, and the fact that out Augmented Dickey-Fuller test resulted in us rejecting the hypothesis that the series is non-stationary, we believe that our current series is suitable for forecasting.

On the basis of the test results, we were able to reject the null hypothesis and conclude that there is good support for the proposition that our new series is stationary. 

In order to ensure we use regression algorithms on our series, we then checked for autocorrelation. We found that the original series was highly autocorrelated. While the differenced series also showed some instances of high autocorrelation, the peaks were far lower than the original series, especially with regard to the partial autocorrelation plot. As a result, we concluded that the differences series is sufficiently uncorrelated for our present purposes.

### Create Features

```python
#Create function to add features to our series.
def featurize(df, lags, mean_window, sd_window):
    #Create time/date features.
    df_diff['hour'] = df_diff.index.hour
    df_diff['day'] = df_diff.index.day
    df_diff['dayofweek'] = df_diff.index.dayofweek
    df_diff['month'] = df_diff.index.month
    
    #Create lag order features.
    for lag in range(1, lags + 1):
        df_diff[f'lag_{lag}'] = df_diff['num_orders'].shift(lag)
        
    #Create rolling mean and standard deviation features.
    #Offset the series by 1, so that current value not considered.
    df_diff['rolling_mean'] = df_diff['num_orders'].shift().rolling(mean_window).mean()
    df_diff['rolling_std'] = df_diff['num_orders'].shift().rolling(sd_window).std()
```

### Train Regression Models

In this section we we will train and test several regression models to determine whether we can predict, with a high enough accuracy, future taxi orders. In doing so, we will be using the training and validation datasets previously created. After each model is trained and run, the RMSE scores for the training and validation sets will be saved in dictionaries, along with the optimal parameters found for each model; these scores will be displayed in the conclusion of this section. 

Prior to training models, we will create a custom class for working with our models. This class will allow us to create a pipeline for standardizing datasets for distance-based algorithms. It also runs each Sklearn model through GridSearchCV in order to determine the optimal hyperparameters. In order to use cross validation on our data, we will create a custom cross validator using TimeSeriesSplit(). We will also create a custom scorer to pass through GridSearchCV; this scorer will return the RMSE score of the model with regard to a specific dataset.

#### Create Class for Models, Scorer, and Cross Validator

```python
#Create function to return RMSE.
def rmse_score(target, prediction):
    return round(math.sqrt(mse(target, prediction)), 2)

#Create a scorer from rmse for GridSearchCV.
rmse = make_scorer(rmse_score, greater_is_better=False)

#Create a cross validator for a time series. 
split = TimeSeriesSplit()
```

```python
#Create class to find best models while using a pipeline to transform data.
class create_model:
    
    #Initialize model with required variables.
    def __init__(self, model, params, name):
        self.model = model
        self.params = params
        self.name = name
        self.rmse = None
        self.valid_score = None
        self.best_params = None
        self.best_rate = None
        self.final_test_score = None
        
    #Function for training and testing model.
    def score_it(self, scale, gradient):
        
        #Command for sklearn models.
        if gradient == False:
            
            #Create pipeline with scaler if non-tree based algorithm.
            if scale == True:
                pipeline = Pipeline(steps=[['scale', StandardScaler()], 
                               ['classifier', self.model]])
            else:
                pipeline = Pipeline(steps=[['classifier', self.model]])
            
            #Use GridSearchCV to tune hyperparameters.    
            grid_search = GridSearchCV(estimator=pipeline, param_grid=self.params, scoring=rmse, cv=split, \
                           n_jobs=-1, verbose = 0, refit=True)
        
            #Fit best model and obtain relevant values.
            grid_search.fit(train_X, train_y)
            self.rmse = -1 * grid_search.best_score_
            self.valid_score = -1 * grid_search.score(validate_X, validate_y)
            self.best_params = grid_search.best_params_
        
        #Command for gradient boosting models.
        else:
            best_rate = 0
            best_valid_rmse = 0

            #Create for loop to tune learning_rate hyperparameter of CatBoostRegressor.
            for i in np.arange(.01, .1, .01):
                reg = self.model
                reg.fit(train_X, train_y)
                self.rmse = rmse_score(train_y, reg.predict(train_X))
                model_valid_score = rmse_score(validate_y, reg.predict(validate_X))
                if best_valid_rmse == 0:
                    best_valid_rmse = model_valid_score
                    best_rate = i
                elif model_valid_score < best_valid_rmse:
                    best_valid_rmse = model_valid_score
                    best_rate = i
                else:
                    continue
            
            #Metrics for gradient descent algorithm.
            self.valid_score = best_valid_rmse
            self.best_rate = best_rate
```

```python
#Create dictionary to store each model's test scores and training time.
dict_training_scores = {}
dict_valid_scores = {}
dict_best_params = {}
```
